## 2. Маркування статі

- Підпишіться на розсилку новин та станьте **першим** хто дізнається про наші акції та розпродажі 
- Також у зоні нетворкінгу **учасники** заходу продемонструють свої AI-розробки: сервіси, послуги та програми у сфері АІ.
- В 2019 році державну підсумкову атестацію у формі ЗНО складатимуть лише **учні** 11-х класів, а також **студенти** закладів вищої (коледжі та технікуми) та професійної освіти, які цього року завершуватимуть здобуття повної загальної середньої освіти.
- Четвертий тест **учасник** ЗНО зможе обрати із загального переліку тестових предметів, що пропонується Українським центром оцінювання якості освіти.
- Як **пацієнти** рятують **лікарів** від коронавірусу, або Де знайти 5 000 доларів на кожну лікарню
- Ви маєте можливість звернутись безпосередньо до кожного **співробітника** Goethe-Institut в Україні.
