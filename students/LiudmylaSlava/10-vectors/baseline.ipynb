```python
import spacy
import gzip
import json
import numpy as np
import pandas as pd
pd.set_option('display.max_colwidth', 1000)
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
uk = spacy.load("/tmp/la_vectors_wiki_lg")

def vec(word):
    return uk(word)[0].vector

def doc_vector(text):
    words = [token.text for token in uk(text) if token.has_vector ]
    return np.mean([vec(w) for w in words], axis=0)
```

```python
def read_file(f):
    with gzip.open(f, 'rt', encoding='utf-8') as inf:
        j = json.load(inf)
        return j[0]['CallZText'], j[0]['CallZType']
read_file('1551.gov.ua/raw/А/А-1025.json.gz')
```

```python
data = []
import glob
for f in glob.glob('1551.gov.ua/raw/*/*.gz'):
    data.append(read_file(f))
data[0]
```

```python
len(data)
```

```python
df = pd.DataFrame(data=data, columns=['text', 'cat'])
df = df.sample(frac = 0.3) 
df.head()
```

```python
def words(text):
    words = [token.text for token in uk(text) if token.has_vector]
    return ' '.join(words)
df['words'] = df.apply(lambda row: words(row['text']),axis=1)
```

```python
df.head()
```

```python
freq = df.groupby('cat', as_index=False)['text'].count().sort_values('text', ascending=False)
f_cat = freq[freq['text']>50]['cat'].tolist()
len(f_cat)
```

```python
df = df[df['cat'].isin(f_cat)]
len(df)
```

```python
df['vect'] = df.apply(lambda row: doc_vector(row['text']),axis=1)
df = df[df['vect'].notnull()]
```

```python
cat_unique = list(set(df['cat'].tolist()))
values = [i for i in range(len(cat_unique))]
d = dict(zip(cat_unique,values))
df['target'] = df['cat'].apply(lambda x : d[x])
df.head()
```

```python
df.head(20)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['vect'].values, df['target'].values, test_size=0.3, random_state=42, stratify=df['target'].values, shuffle=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=5)
```

```python
KNN.fit(list(X_train), y_train)
```

```python
%%time
predicted = KNN.predict(list(X_test))
print(classification_report(y_test, predicted))
```

```python
from numpy import dot
from numpy.linalg import norm

def cos_sim(a, b):
    return dot(a, b)/(norm(a)*norm(b))
```
