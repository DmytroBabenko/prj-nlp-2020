# 3. Використання довільних алгоритмів класифікації

```python
from utils import load_data
import pandas as pd
```

```python
data = load_data()
```

```python
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

```python
def run_model(model, data):
    x, y = data.values()
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
    
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)

    return pd.DataFrame(classification_report(y_test, y_pred, output_dict=True))
```

## 3.1 LogisticRegression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(C=1, solver='saga', n_jobs=-1, class_weight='balanced')
run_model(model, data)
```

## 3.2 RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_jobs=-1, class_weight='balanced')
run_model(model, data)
```

## 3.3 SVM 

```python
from sklearn.svm import SVC

model = SVC(n_jobs=-1, class_weight='balanced')
run_model(model, data)
```
